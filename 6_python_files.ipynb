## 6. Working with Files in Python

In JavaScript (Node.js), you use the `fs` module to read, write, and manage files. In Python, most of this is **built in** — no imports needed for basic file I/O. For file system operations (rename, delete, etc.), you use the `os` and `shutil` modules.

| What you want | JavaScript (Node.js) | Python |
|---|---|---|
| Read a file | `fs.readFileSync('f.txt', 'utf8')` | `open('f.txt').read()` |
| Write a file | `fs.writeFileSync('f.txt', data)` | `open('f.txt', 'w').write(data)` |
| Append to a file | `fs.appendFileSync('f.txt', data)` | `open('f.txt', 'a').write(data)` |
| Check if file exists | `fs.existsSync('f.txt')` | `os.path.exists('f.txt')` |
| Delete a file | `fs.unlinkSync('f.txt')` | `os.remove('f.txt')` |
| Rename / move | `fs.renameSync('old', 'new')` | `os.rename('old', 'new')` |
| Create a directory | `fs.mkdirSync('dir')` | `os.makedirs('dir')` |
| List directory | `fs.readdirSync('dir')` | `os.listdir('dir')` |
| Delete a directory | `fs.rmSync('dir', {recursive: true})` | `shutil.rmtree('dir')` |

Key differences:
- Python uses the **`with` statement** to safely open files — it automatically closes the file when done (like a try/finally)
- File modes: `'r'` (read), `'w'` (write — overwrites!), `'a'` (append), `'x'` (create — fails if exists)
- Python's `os` and `pathlib` modules handle paths, while Node uses `path`
- No need to specify encoding for text files — Python 3 defaults to UTF-8

### Examples — Opening and reading files

```python
# The `with` statement — Python's recommended way to work with files
# It automatically closes the file when the block ends, even if an error occurs
# JS equivalent: try { const f = fs.openSync(...) } finally { fs.closeSync(f) }

# First, let's create a sample file to work with
with open("sample.txt", "w") as f:
    f.write("Line 1: Hello, Python!\n")
    f.write("Line 2: File I/O is easy.\n")
    f.write("Line 3: This is the last line.")

print("File created!")
```

```python
# Reading the entire file at once — like fs.readFileSync()

with open("sample.txt", "r") as f:
    content = f.read()

print(content)
print()
print(f"Total characters: {len(content)}")
```

```python
# Reading line by line — very useful for large files
# No JS equivalent that's this clean

with open("sample.txt", "r") as f:
    for line_number, line in enumerate(f, start=1):
        print(f"{line_number}: {line}", end="")  # lines already end with \n

print()  # blank line

# Or read all lines into a list — like content.split('\n') in JS
with open("sample.txt", "r") as f:
    lines = f.readlines()

print(f"Number of lines: {len(lines)}")
print(f"First line: {lines[0].strip()}")  # .strip() removes the trailing \n
```

### Examples — Writing and appending

```python
# Writing a file — mode 'w' creates the file or OVERWRITES it entirely
# JS: fs.writeFileSync('output.txt', data)

students = ["Ada", "Bob", "Cat", "Dan"]

with open("students.txt", "w") as f:
    for student in students:
        f.write(f"{student}\n")    # write() does NOT add a newline automatically

# Verify
with open("students.txt", "r") as f:
    print(f.read())
```

```python
# Appending to a file — mode 'a' adds to the end without erasing existing content
# JS: fs.appendFileSync('students.txt', data)

with open("students.txt", "a") as f:
    f.write("Eve\n")
    f.write("Fay\n")

# Verify — Eve and Fay are added at the end
with open("students.txt", "r") as f:
    print(f.read())
```

```python
# Exclusive creation — mode 'x' creates a new file but FAILS if it already exists
# Useful to prevent accidentally overwriting data

try:
    with open("students.txt", "x") as f:
        f.write("This won't work!")
except FileExistsError:
    print("File already exists! Use 'w' to overwrite or 'a' to append.")
```

### Examples — File system operations

```python
import os

# Check if a file or directory exists
# JS: fs.existsSync('sample.txt')

print("sample.txt exists:", os.path.exists("sample.txt"))     # True
print("ghost.txt exists:", os.path.exists("ghost.txt"))       # False

# Check specifically for file vs directory
print("Is a file:", os.path.isfile("sample.txt"))             # True
print("Is a directory:", os.path.isdir("sample.txt"))         # False
```

```python
import os

# Renaming a file
# JS: fs.renameSync('students.txt', 'class_roster.txt')

os.rename("students.txt", "class_roster.txt")
print("Renamed! Files now:", [f for f in os.listdir(".") if f.endswith(".txt")])

# Rename it back for later examples
os.rename("class_roster.txt", "students.txt")
```

```python
import os

# Deleting a file
# JS: fs.unlinkSync('sample.txt')

os.remove("sample.txt")
print("sample.txt deleted:", not os.path.exists("sample.txt"))  # True

# Trying to delete a file that doesn't exist raises an error
try:
    os.remove("ghost.txt")
except FileNotFoundError:
    print("Can't delete ghost.txt — file not found!")
```

### Examples — Working with directories

```python
import os

# Creating directories
# JS: fs.mkdirSync('data/reports', { recursive: true })

# os.makedirs creates the full path, including parent directories
os.makedirs("data/reports", exist_ok=True)   # exist_ok=True prevents error if it exists
print("Created data/reports/")

# List contents of a directory
# JS: fs.readdirSync('.')
print("\nFiles in current directory:")
for item in sorted(os.listdir(".")):
    kind = "DIR " if os.path.isdir(item) else "FILE"
    print(f"  {kind}  {item}")
```

```python
import os
import shutil

# Deleting a directory
# os.rmdir() only works on EMPTY directories
# shutil.rmtree() deletes everything recursively — like rm -rf
# JS: fs.rmSync('data', { recursive: true })

# Let's create a file inside the directory first
with open("data/reports/summary.txt", "w") as f:
    f.write("Q1 Report")

# This would fail because the directory is not empty:
# os.rmdir("data/reports")  → OSError

# Use shutil.rmtree to delete everything
shutil.rmtree("data")
print("data/ directory deleted:", not os.path.exists("data"))
```

### Examples — Copying and moving files

```python
import shutil
import os

# Copying files
# JS: fs.copyFileSync('students.txt', 'students_backup.txt')

shutil.copy("students.txt", "students_backup.txt")
print("Copied! Both files exist:")
print("  students.txt:", os.path.exists("students.txt"))
print("  students_backup.txt:", os.path.exists("students_backup.txt"))

# Moving a file — like rename but can move across directories
os.makedirs("archive", exist_ok=True)
shutil.move("students_backup.txt", "archive/students_backup.txt")
print("\nMoved to archive/")
print("  archive contents:", os.listdir("archive"))

# Clean up
shutil.rmtree("archive")
```

### Examples — Practical patterns

```python
import os

# Pattern: safe file writing — check before overwriting

def safe_write(filename, content, overwrite=False):
    if os.path.exists(filename) and not overwrite:
        print(f"'{filename}' already exists. Use overwrite=True to replace it.")
        return False
    with open(filename, "w") as f:
        f.write(content)
    print(f"Written to '{filename}'")
    return True

safe_write("students.txt", "new content")               # Already exists!
safe_write("students.txt", "new content", overwrite=True)  # Overwrites
safe_write("new_file.txt", "hello!")                     # Creates new file

# Clean up
os.remove("new_file.txt")
```
