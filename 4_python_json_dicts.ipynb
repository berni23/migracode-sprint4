## Dictionaries & JSON in Python

In JavaScript you have **objects** (`{}`). In Python, the equivalent is a **dictionary** (`dict`). They work very similarly, but with a few important differences.

| What you want | JavaScript | Python |
|---|---|---|
| Create | `const obj = {name: "Ada"}` | `d = {"name": "Ada"}` |
| Access value | `obj.name` or `obj["name"]` | `d["name"]` or `d.get("name")` |
| Set value | `obj.age = 25` | `d["age"] = 25` |
| Delete key | `delete obj.age` | `del d["age"]` |
| Check key exists | `"name" in obj` | `"name" in d` |
| Get all keys | `Object.keys(obj)` | `d.keys()` |
| Get all values | `Object.values(obj)` | `d.values()` |
| Get key-value pairs | `Object.entries(obj)` | `d.items()` |
| Merge | `{...obj1, ...obj2}` | `{**d1, **d2}` or `d1 \| d2` |
| Parse JSON string | `JSON.parse(str)` | `json.loads(str)` |
| Convert to JSON string | `JSON.stringify(obj)` | `json.dumps(d)` |

Key differences:
- Python dict keys **must be quoted** — `{"name": "Ada"}` not `{name: "Ada"}`
- No dot notation for access — use `d["name"]` or `d.get("name")`
- `.get(key, default)` returns a default instead of raising an error if the key is missing
- Python uses `True`/`False`/`None` instead of `true`/`false`/`null`

### Examples — Dictionary basics

```python
# Creating a dictionary
student = {
    "name": "Ada",
    "age": 25,
    "courses": ["Python 101", "Data Science"],
    "active": True
}

# Accessing values
print(student["name"])            # Ada
print(student.get("age"))         # 25
print(student.get("gpa", 0.0))   # 0.0 — key doesn't exist, returns default

# Adding / updating
student["email"] = "ada@example.com"
student["age"] = 26
print(student)
```

```python
# Iterating over a dictionary
profile = {"name": "Luna", "role": "student", "city": "Barcelona"}

# Loop over keys (default)
for key in profile:
    print(key, "->", profile[key])

print()

# Loop over key-value pairs — like Object.entries() in JS
for key, value in profile.items():
    print(f"{key}: {value}")
```

```python
# Merging dictionaries
defaults = {"theme": "light", "lang": "en", "notifications": True}
user_prefs = {"theme": "dark", "font_size": 14}

# Using ** unpacking (like JS spread)
merged = {**defaults, **user_prefs}
print(merged)
# {'theme': 'dark', 'lang': 'en', 'notifications': True, 'font_size': 14}

# Using | operator (Python 3.9+)
merged2 = defaults | user_prefs
print(merged2)
```

```python
# Dictionary comprehensions — like list comprehensions but for dicts
# JS: Object.fromEntries(arr.map(x => [x, x.length]))

words = ["hello", "world", "python"]
word_lengths = {w: len(w) for w in words}
print(word_lengths)  # {'hello': 5, 'world': 5, 'python': 6}

# Filtering a dictionary
scores = {"Ada": 92, "Bob": 45, "Cat": 78, "Dan": 31}
passing = {name: score for name, score in scores.items() if score >= 50}
print(passing)  # {'Ada': 92, 'Cat': 78}
```

### Examples — Working with JSON

```python
import json

# --- Parsing a JSON string into a Python dict ---
# JS: const data = JSON.parse(jsonString)

json_string = '{"name": "Ada", "age": 25, "courses": ["Python", "Data Science"]}'
data = json.loads(json_string)

print(type(data))       # <class 'dict'>
print(data["name"])     # Ada
print(data["courses"])  # ['Python', 'Data Science']
```

```python
import json

# --- Converting a dict to a JSON string ---
# JS: const str = JSON.stringify(obj, null, 2)

student = {
    "name": "Ada",
    "age": 25,
    "graduated": False,
    "gpa": None
}

# Compact
print(json.dumps(student))
# {"name": "Ada", "age": 25, "graduated": false, "gpa": null}
#                             ^^^^^ Python's False → JSON false
#                                                    ^^^^ Python's None → JSON null

print()

# Pretty-printed
print(json.dumps(student, indent=2))
```

```python
import json

# --- Reading and writing JSON files ---
# This is something you'd do a lot in real projects

data = {
    "students": [
        {"name": "Ada", "grade": "A"},
        {"name": "Bob", "grade": "B"}
    ]
}

# Write to file — like fs.writeFileSync('data.json', JSON.stringify(data, null, 2))
with open("sample_data.json", "w") as f:
    json.dump(data, f, indent=2)

print("File written!")

# Read from file — like JSON.parse(fs.readFileSync('data.json'))
with open("sample_data.json", "r") as f:
    loaded = json.load(f)

print(loaded)
print(loaded["students"][0]["name"])  # Ada
```

### Examples — Nested data manipulation

Real-world JSON is almost always nested. Here's how to navigate and transform it.

```python
import json

# Imagine this came from an API response
api_response = '''
{
  "status": "ok",
  "results": [
    {"id": 1, "title": "Learn Python", "completed": false, "tags": ["education", "coding"]},
    {"id": 2, "title": "Buy groceries", "completed": true, "tags": ["personal"]},
    {"id": 3, "title": "Build portfolio", "completed": false, "tags": ["coding", "career"]},
    {"id": 4, "title": "Read a book", "completed": true, "tags": ["personal", "education"]},
    {"id": 5, "title": "Practice algorithms", "completed": false, "tags": ["coding"]}
  ]
}
'''

data = json.loads(api_response)

# Get just the results list
todos = data["results"]

# Filter: only incomplete todos
# JS: todos.filter(t => !t.completed)
pending = [t for t in todos if not t["completed"]]
print("Pending:")
for t in pending:
    print(f"  - {t['title']}")

print()

# Extract: just the titles of completed todos
# JS: todos.filter(t => t.completed).map(t => t.title)
done_titles = [t["title"] for t in todos if t["completed"]]
print("Completed:", done_titles)

print()

# Collect all unique tags across all todos
# JS: [...new Set(todos.flatMap(t => t.tags))]
all_tags = set()
for t in todos:
    all_tags.update(t["tags"])
print("All tags:", sorted(all_tags))
```

```python
import json

# Transforming data structure — a very common task
# Convert a list of records into a lookup dictionary keyed by ID

users = [
    {"id": "u1", "name": "Ada", "role": "admin"},
    {"id": "u2", "name": "Bob", "role": "editor"},
    {"id": "u3", "name": "Cat", "role": "viewer"}
]

# JS: Object.fromEntries(users.map(u => [u.id, u]))
users_by_id = {u["id"]: u for u in users}

print(json.dumps(users_by_id, indent=2))

# Now you can look up any user by ID in O(1)
print()
print("User u2:", users_by_id["u2"]["name"])  # Bob
```

### Shallow copy vs Deep copy

In JavaScript, spreading an object (`{...obj}`) only copies the **top level** — nested objects still point to the same reference. Python works the same way.

| What you want | JavaScript | Python |
|---|---|---|
| Shallow copy | `{...obj}` or `Object.assign({}, obj)` | `d.copy()` or `{**d}` |
| Deep copy | `structuredClone(obj)` | `copy.deepcopy(d)` |

**Shallow copy** — creates a new dict, but nested objects (lists, dicts) are still **shared** references.
**Deep copy** — creates a completely independent clone, including all nested structures.

```python
# --- Shallow copy: nested objects are SHARED ---

original = {
    "name": "Ada",
    "scores": [90, 85, 78]
}

shallow = original.copy()  # or: shallow = {**original}

# Changing a top-level key does NOT affect the original
shallow["name"] = "Bob"
print("original name:", original["name"])  # Ada  ✓ independent
print("shallow name: ", shallow["name"])   # Bob

# But mutating a nested list DOES affect the original!
shallow["scores"].append(100)
print()
print("original scores:", original["scores"])  # [90, 85, 78, 100]  ← surprise!
print("shallow scores: ", shallow["scores"])   # [90, 85, 78, 100]
```

```python
# --- Deep copy: everything is fully independent ---
import copy

original = {
    "name": "Ada",
    "scores": [90, 85, 78]
}

deep = copy.deepcopy(original)

# Mutating nested data does NOT affect the original
deep["scores"].append(100)

print("original scores:", original["scores"])  # [90, 85, 78]  ✓ safe!
print("deep scores:    ", deep["scores"])      # [90, 85, 78, 100]
```

```python
# --- Quick trick: json.loads(json.dumps()) as a deep copy ---
# If your dict only contains JSON-compatible types (strings, numbers, bools, None, lists, dicts),
# you can deep-copy without importing copy:
import json

original = {"name": "Ada", "scores": [90, 85, 78]}
clone = json.loads(json.dumps(original))

clone["scores"].append(100)
print("original scores:", original["scores"])  # [90, 85, 78]  ✓ safe
print("clone scores:   ", clone["scores"])     # [90, 85, 78, 100]
```
