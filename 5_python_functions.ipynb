## 5. Functions in Python

In JavaScript you define functions with `function` or arrow syntax `() => {}`. In Python you use the **`def`** keyword. The concepts are almost identical — but the syntax and some features are quite different.

| What you want | JavaScript | Python |
|---|---|---|
| Define a function | `function greet(name) { ... }` | `def greet(name):` |
| Arrow / lambda | `const double = x => x * 2` | `double = lambda x: x * 2` |
| Default parameter | `function greet(name = "World")` | `def greet(name="World"):` |
| Return a value | `return result` | `return result` |
| Return multiple values | `return [a, b]` or `return {a, b}` | `return a, b` (tuple) |
| Rest parameters | `function f(...args)` | `def f(*args):` |
| Keyword arguments | _(object destructuring)_ | `def f(**kwargs):` |
| Docstrings | `/** JSDoc */` | `"""docstring"""` |
| First-class functions | `const fn = greet` | `fn = greet` |
| Callback / higher-order | `arr.map(fn)` | `map(fn, arr)` or comprehension |

Key differences:
- **No braces** — the function body is defined by **indentation** (just like loops and if statements)
- **No hoisting** — you must define a function **before** you call it
- Python functions can return **multiple values** as a tuple — no need to wrap them in an array or object
- **`*args`** and **`**kwargs`** give you flexible arguments (more powerful than JS rest/spread)
- **Lambda** functions are limited to a single expression (no multi-line arrow functions)

### Examples — Defining and calling functions

```python
# Basic function definition
# JS: function greet(name) { return `Hello, ${name}!`; }

def greet(name):
    return f"Hello, {name}!"

print(greet("Ada"))       # Hello, Ada!
print(greet("Python"))    # Hello, Python!
```

```python
# Functions without a return statement return None (like undefined in JS)

def say_hello(name):
    print(f"Hello, {name}!")

result = say_hello("Bob")
print(f"Return value: {result}")   # Return value: None
```

```python
# Default parameters — just like JS, but Python had them first!
# JS: function greet(name = "World", greeting = "Hello")

def greet(name="World", greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet())                          # Hello, World!
print(greet("Ada"))                     # Hello, Ada!
print(greet("Ada", "Hi"))               # Hi, Ada!
print(greet(greeting="Hey", name="Bob")) # Hey, Bob!  ← keyword arguments!
```

```python
# Keyword arguments — call with name=value in any order
# This has no direct equivalent in JS (closest is passing an options object)

def create_profile(name, age, city="Unknown", role="student"):
    return {"name": name, "age": age, "city": city, "role": role}

# All of these work:
print(create_profile("Ada", 25))
print(create_profile("Ada", 25, city="Barcelona"))
print(create_profile("Ada", 25, role="admin", city="Barcelona"))
print(create_profile(name="Ada", age=25))   # even positional args can be named
```

```python
# Returning multiple values — Python's superpower!
# In JS you'd return an array or object: return { min: ..., max: ..., avg: ... }

def get_stats(numbers):
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

# Unpack directly into variables (like JS destructuring)
lo, hi, avg = get_stats([10, 20, 30, 40, 50])
print(f"Min: {lo}, Max: {hi}, Avg: {avg}")

# Or keep it as a tuple
result = get_stats([10, 20, 30, 40, 50])
print(f"Result tuple: {result}")
print(f"Type: {type(result)}")   # <class 'tuple'>
```

### Examples — `*args` and `**kwargs`

```python
# *args — collect extra positional arguments into a tuple
# JS: function sum(...numbers) { return numbers.reduce((a, b) => a + b, 0); }

def add_all(*numbers):
    print(f"Received: {numbers}")    # it's a tuple
    return sum(numbers)

print(add_all(1, 2, 3))          # 6
print(add_all(10, 20, 30, 40))   # 100
```

```python
# **kwargs — collect extra keyword arguments into a dictionary
# JS closest equivalent: function config(options) { ... } with an object

def build_url(base, **params):
    query = "&".join(f"{k}={v}" for k, v in params.items())
    return f"{base}?{query}" if query else base

print(build_url("https://api.example.com/search", q="python", page=1, limit=10))
# https://api.example.com/search?q=python&page=1&limit=10
```

```python
# Combining *args and **kwargs — accepts ANYTHING
# This pattern is very common in decorators and wrappers

def debug_call(func_name, *args, **kwargs):
    print(f"Calling {func_name}")
    print(f"  Positional args: {args}")
    print(f"  Keyword args: {kwargs}")

debug_call("save_user", "Ada", 25, role="admin", active=True)
```

### Examples — Lambda functions

```python
# Lambda — anonymous single-expression functions
# JS: const double = x => x * 2
# Python: lambda x: x * 2

double = lambda x: x * 2
print(double(5))   # 10

# Lambdas are most useful as arguments to other functions
# JS: names.sort((a, b) => a.length - b.length)

names = ["Charlie", "Ada", "Bob", "Delilah"]
names.sort(key=lambda name: len(name))
print(names)   # ['Ada', 'Bob', 'Charlie', 'Delilah']

# Sorting a list of dicts by a specific key
students = [
    {"name": "Bob", "grade": 72},
    {"name": "Ada", "grade": 95},
    {"name": "Cat", "grade": 88},
]
students.sort(key=lambda s: s["grade"], reverse=True)
print(students)   # sorted by grade descending
```

### Examples — Functions as first-class citizens

```python
# Functions are values — you can pass them around, just like in JS

def shout(text):
    return text.upper() + "!!!"

def whisper(text):
    return text.lower() + "..."

def speak(text, style_func):
    return style_func(text)

print(speak("Hello World", shout))     # HELLO WORLD!!!
print(speak("Hello World", whisper))   # hello world...
```

```python
# map(), filter(), and reduce() — just like JS array methods but as standalone functions
# (In practice, Python devs prefer list comprehensions over map/filter)

numbers = [1, 2, 3, 4, 5]

# map — apply a function to each element
# JS: numbers.map(x => x ** 2)
squared = list(map(lambda x: x ** 2, numbers))
print("map:", squared)        # [1, 4, 9, 16, 25]

# filter — keep elements that return True
# JS: numbers.filter(x => x % 2 === 0)
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("filter:", evens)       # [2, 4]

# The Pythonic way — list comprehensions (usually preferred)
squared_lc = [x ** 2 for x in numbers]
evens_lc = [x for x in numbers if x % 2 == 0]
print("comprehension (squared):", squared_lc)
print("comprehension (evens):", evens_lc)
```

### Examples — Scope and closures

```python
# Scope works similarly to JS — local variables are local

x = "global"

def show_scope():
    x = "local"             # this is a NEW local variable, doesn't affect the global
    print(f"Inside: {x}")   # local

show_scope()
print(f"Outside: {x}")     # global
```

```python
# Closures — functions that remember their enclosing scope
# This works exactly like JS closures

# JS: function makeCounter() { let count = 0; return () => ++count; }

def make_counter():
    count = 0
    def increment():
        nonlocal count       # needed to modify the outer variable (like let in JS closure)
        count += 1
        return count
    return increment

counter = make_counter()
print(counter())   # 1
print(counter())   # 2
print(counter())   # 3

# Each counter is independent
counter2 = make_counter()
print(counter2())  # 1  — starts fresh
```

### Examples — Docstrings

```python
# Docstrings — built-in documentation for your functions
# Like JSDoc, but it's actually part of the language

def calculate_bmi(weight_kg, height_m):
    """Calculate Body Mass Index (BMI).

    Args:
        weight_kg: Weight in kilograms.
        height_m: Height in meters.

    Returns:
        The BMI value as a float.
    """
    return weight_kg / (height_m ** 2)

print(calculate_bmi(70, 1.75))

# You can access the docstring at runtime!
print()
print(calculate_bmi.__doc__)

# Or use the built-in help()
help(calculate_bmi)
```
