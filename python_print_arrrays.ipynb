## 1. Printing in Python

In JavaScript, you're used to `console.log()` to output things to the console. In Python, the equivalent is the **`print()`** function — but it comes with a few more tricks out of the box.

| What you want | JavaScript | Python |
|---|---|---|
| Print a value | `console.log("hello")` | `print("hello")` |
| Print multiple values | `console.log("a", "b", "c")` | `print("a", "b", "c")` |
| Print without newline | _(tricky)_ | `print("hello", end="")` |
| Print with custom separator | _(manual)_ | `print("a", "b", "c", sep=" - ")` |
| String interpolation | `` `Hello ${name}` `` | `f"Hello {name}"` |

Key differences:
- `print()` adds a **newline at the end** by default (you can change this with `end=""`)
- `print()` separates multiple arguments with a **space** by default (you can change this with `sep=`)
- Python's **f-strings** (`f"..."`) are the equivalent of JS template literals (`` `...` ``)

### Examples

```python
# Basic print — just like console.log()
print("Hello, Python!")

# Print multiple values — they get separated by a space
print("My name is", "Ada", "and I am", 25, "years old")
```

```python
# Custom separator — instead of spaces between values
print("2026", "03", "25", sep="-")       # Output: 2026-03-25
print("apple", "banana", "cherry", sep=" | ")
```

```python
# Controlling the end character — by default print() adds a newline (\n)
print("Loading", end="")
print("...", end="")
print(" Done!")    # Output: Loading... Done!  (all on one line)
```

```python
# f-strings — Python's template literals
# In JS:  `Hello ${name}, you are ${age} years old`
# In Python:

name = "Ada"
age = 25
print(f"Hello {name}, you are {age} years old")

# You can put any expression inside the { }
print(f"In 5 years you'll be {age + 5}")
print(f"Your name in uppercase: {name.upper()}")
```

```python
# Other ways to format strings (older styles, good to recognise)

# .format() method
print("Hello {}, you are {} years old".format("Ada", 25))

# % formatting (oldest style — you'll see it in legacy code)
print("Hello %s, you are %d years old" % ("Ada", 25))
```

### Bonus: Pretty-printing JSON

In JavaScript you might do `JSON.stringify(obj, null, 2)` to get a nicely formatted JSON output. Python has an equivalent with the built-in `json` module.

```python
import json

student = {
    "name": "Ada",
    "age": 25,
    "courses": ["Python 101", "Data Science"],
    "address": {
        "city": "Barcelona",
        "country": "Spain"
    }
}

# Without pretty-print — hard to read (like console.log(obj) in JS)
print(student)

print()  # blank line

# Pretty-print with json.dumps()  —  like JSON.stringify(obj, null, 2) in JS
print(json.dumps(student, indent=2))
```

## 2. Lists — Python's Arrays

In JavaScript you have **Arrays**. In Python, the equivalent is a **list**. They look almost identical, but Python lists come with some extra superpowers — especially **slicing** and **list comprehensions**.

| What you want | JavaScript | Python |
|---|---|---|
| Create | `const arr = [1, 2, 3]` | `arr = [1, 2, 3]` |
| Access element | `arr[0]` | `arr[0]` |
| Last element | `arr[arr.length - 1]` | `arr[-1]` |
| Length | `arr.length` | `len(arr)` |
| Add to end | `arr.push(4)` | `arr.append(4)` |
| Remove last | `arr.pop()` | `arr.pop()` |
| Remove by index | `arr.splice(i, 1)` | `del arr[i]` or `arr.pop(i)` |
| Check if exists | `arr.includes(2)` | `2 in arr` |
| Slice | `arr.slice(1, 3)` | `arr[1:3]` |
| Concatenate | `[...arr1, ...arr2]` | `arr1 + arr2` |
| Sort | `arr.sort()` | `arr.sort()` or `sorted(arr)` |
| Reverse | `arr.reverse()` | `arr.reverse()` or `arr[::-1]` |
| Map | `arr.map(x => x * 2)` | `[x * 2 for x in arr]` |
| Filter | `arr.filter(x => x > 2)` | `[x for x in arr if x > 2]` |

Key differences:
- **Negative indexing**: `arr[-1]` gets the last element, `arr[-2]` the second to last, etc. No more `arr[arr.length - 1]`!
- **Slicing** `arr[start:end]` returns a new list — cleaner than `.slice()`
- **List comprehensions** replace `.map()` and `.filter()` — more on this below
- `len()` is a **function**, not a property (no `.length`)

### Examples

```python
# Creating and accessing lists
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

print(fruits[0])     # First element → apple
print(fruits[-1])    # Last element → elderberry
print(fruits[-2])    # Second to last → date
print(len(fruits))   # Length → 5
```

```python

myArray = [2,3,4,5,6]
#get last element
print(myArray[-1])
#append
myArray2 = myArray
myArray.append(2)
print(myArray)
```

```python
# Slicing — arr[start:end] (end is exclusive, just like JS .slice())
numbers = [10, 20, 30, 40, 50, 60, 70]

print(numbers[1:4])    # [20, 30, 40]  — from index 1 up to (not including) 4
print(numbers[:3])     # [10, 20, 30]  — from the start up to index 3
print(numbers[4:])     # [50, 60, 70]  — from index 4 to the end
print(numbers[::2])    # [10, 30, 50, 70] — every 2nd element (step = 2)
print(numbers[::-1])   # [70, 60, 50, 40, 30, 20, 10] — reversed!
```

```python
# Modifying lists
colors = ["red", "green", "blue"]

colors.append("yellow")        # JS: colors.push("yellow")
print("After append:", colors)

colors.insert(1, "orange")     # Insert at index 1
print("After insert:", colors)

removed = colors.pop()         # Removes and returns last element
print(f"Removed: {removed}, List: {colors}")

colors.pop(0)                  # Remove by index (like splice)
print("After pop(0):", colors)

del colors[1]                  # Another way to remove by index
print("After del [1]:", colors)
```

```python
# Checking membership — much cleaner than JS .includes()
animals = ["cat", "dog", "rabbit", "hamster"]

print("cat" in animals)        # True   (JS: animals.includes("cat"))
print("snake" in animals)      # False
print("snake" not in animals)  # True   — there's even a "not in"!
```

```python
# Sorting
scores = [42, 17, 93, 8, 56]

# sorted() returns a NEW list (original unchanged) — like JS [...arr].sort()
print("sorted():", sorted(scores))
print("original:", scores)

# .sort() modifies in place (like JS .sort(), but works correctly with numbers!)
scores.sort()
print("after .sort():", scores)

# Descending order
print("descending:", sorted(scores, reverse=True))
```

```python
# List comprehensions — Python's replacement for .map() and .filter()

numbers = [1, 2, 3, 4, 5]

# .map() equivalent — transform each element
# JS: numbers.map(x => x * 2)
doubled = [x * 2 for x in numbers]
print("Doubled:", doubled)

# .filter() equivalent — keep elements that match a condition
# JS: numbers.filter(x => x > 2)
big = [x for x in numbers if x > 2]
print("Greater than 2:", big)

# .map() + .filter() combined!
# JS: numbers.filter(x => x % 2 === 0).map(x => x ** 2)
even_squares = [x ** 2 for x in numbers if x % 2 == 0]
print("Squares of even numbers:", even_squares)
```

```python
# Useful built-in functions that work with lists
nums = [10, 20, 30, 40, 50]

print("Sum:", sum(nums))          # No JS equivalent (you'd use .reduce())
print("Min:", min(nums))          # JS: Math.min(...nums)
print("Max:", max(nums))          # JS: Math.max(...nums)
print("Any > 25?", any(x > 25 for x in nums))   # JS: nums.some(x => x > 25)
print("All > 5?", all(x > 5 for x in nums))     # JS: nums.every(x => x > 5)
```
