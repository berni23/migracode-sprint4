## 1. The `for` Loop

In JavaScript you have several ways to loop: `for`, `for...of`, `for...in`, `.forEach()`. In Python there's really just **one `for` loop** — and it works like JavaScript's `for...of`.

| What you want | JavaScript | Python |
|---|---|---|
| Loop over items | `for (const item of arr)` | `for item in arr:` |
| Loop with index | `for (let i = 0; i < n; i++)` | `for i in range(n):` |
| Loop with index + item | `arr.forEach((item, i) => ...)` | `for i, item in enumerate(arr):` |
| Loop over object keys | `for (const key in obj)` | `for key in dict:` |
| Loop over key-value pairs | `Object.entries(obj).forEach(...)` | `for key, val in dict.items():` |
| Loop over two lists | _(manual index)_ | `for a, b in zip(list1, list2):` |

Key differences:
- Python `for` **always iterates over something** — a list, a string, a range, a dictionary, etc.
- No parentheses `()` around the expression, but a **colon** `:` at the end
- The loop body is defined by **indentation** (no `{ }` braces)
- `range()` generates a sequence of numbers — it replaces the classic `for (let i = 0; i < n; i++)`

### Examples

````python
# Looping over a list — like JS: for (const fruit of fruits)
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
````

````python
# Looping over a string — each character is an item (same concept in JS)
for char in "Python":
    print(char, end=" ")
# Output: P y t h o n
````

````python
# range() — replaces the classic for (let i = 0; i < n; i++)

# range(stop) — from 0 up to (not including) stop
# JS: for (let i = 0; i < 5; i++)
print("range(5):")
for i in range(5):
    print(i, end=" ")    # 0 1 2 3 4

print()

# range(start, stop) — from start up to (not including) stop
# JS: for (let i = 2; i < 7; i++)
print("range(2, 7):")
for i in range(2, 7):
    print(i, end=" ")    # 2 3 4 5 6

print()

# range(start, stop, step) — with a custom step
# JS: for (let i = 0; i < 20; i += 3)
print("range(0, 20, 3):")
for i in range(0, 20, 3):
    print(i, end=" ")    # 0 3 6 9 12 15 18

print()

# Counting backwards!
# JS: for (let i = 5; i > 0; i--)
print("range(5, 0, -1):")
for i in range(5, 0, -1):
    print(i, end=" ")    # 5 4 3 2 1
````

````python
# enumerate() — loop with index AND value at the same time
# JS: fruits.forEach((fruit, index) => console.log(index, fruit))

fruits = ["apple", "banana", "cherry", "date"]

for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

print()

# You can set a custom start number for the index
for position, fruit in enumerate(fruits, start=1):
    print(f"#{position} → {fruit}")
````

````python
# Looping over dictionaries
# JS: for (const key in obj) / Object.entries(obj).forEach(...)

student = {"name": "Ada", "age": 25, "city": "Barcelona"}

# Loop over keys (default behaviour)
print("Keys:")
for key in student:
    print(f"  {key}")

print()

# Loop over values
print("Values:")
for value in student.values():
    print(f"  {value}")

print()

# Loop over key-value pairs — the most useful one
print("Key-value pairs:")
for key, value in student.items():
    print(f"  {key} → {value}")
````

````python
# zip() — loop over multiple lists in parallel
# In JS you'd have to use an index: for (let i = 0; i < names.length; i++)

names = ["Ada", "Bob", "Charlie"]
scores = [95, 82, 78]
grades = ["A", "B+", "C+"]

for name, score, grade in zip(names, scores, grades):
    print(f"{name} scored {score} ({grade})")
````

````python
# break and continue — work exactly the same as in JavaScript

# break — exit the loop early
print("break example:")
for num in range(10):
    if num == 5:
        print("Found 5, stopping!")
        break
    print(num, end=" ")

print()
print()

# continue — skip the current iteration
print("continue example (skip even numbers):")
for num in range(10):
    if num % 2 == 0:
        continue
    print(num, end=" ")    # 1 3 5 7 9
````

````python
# Nested loops — same concept as in JS, just different syntax

# Print a multiplication table (3x3)
for row in range(1, 4):
    for col in range(1, 4):
        print(f"{row * col:4}", end="")   # :4 pads to 4 characters wide
    print()   # newline after each row
````

## 2. The `while` Loop

The `while` loop works almost identically to JavaScript's — it keeps running **as long as the condition is `True`**. The syntax is simpler though: no parentheses, just a colon.

```python
# JavaScript                    # Python
while (condition) {              while condition:
    // do something                  # do something
}
```

| What you want | JavaScript | Python |
|---|---|---|
| Basic while | `while (x > 0) { ... }` | `while x > 0:` |
| Infinite loop | `while (true) { ... }` | `while True:` |
| Break out | `break` | `break` |
| Skip iteration | `continue` | `continue` |
| Do-while | `do { ... } while (cond)` | _(no equivalent — use `while True` + `break`)_ |

Key differences:
- No parentheses around the condition
- Colon `:` at the end, indentation for the body
- `True` / `False` are capitalised in Python (not `true` / `false`)
- Python has **no `do...while`** — you simulate it with `while True:` and a `break`

### Examples

````python
# Basic while loop — countdown
# JS: let count = 5; while (count > 0) { console.log(count); count--; }

count = 5
while count > 0:
    print(count, end=" ")
    count -= 1          # Python has no count-- or count++

print("Liftoff!")
````

````python
# Accumulating a result — summing numbers until a limit
# "Add numbers 1, 2, 3... until the total exceeds 50"

total = 0
number = 1

while total <= 50:
    total += number
    number += 1

print(f"Stopped at number {number - 1}, total = {total}")
````

````python
# while True + break — Python's version of do...while
# This pattern is very common: loop forever, break when done

# JS equivalent:
# do {
#     answer = prompt("Guess the number");
# } while (answer !== "7");

import random

secret = random.randint(1, 10)
attempts = 0

while True:
    attempts += 1
    guess = random.randint(1, 10)   # simulating a guess (in real life you'd use input())
    print(f"Attempt {attempts}: guessed {guess}", end="")

    if guess == secret:
        print(" ✓ Correct!")
        break
    else:
        print(" ✗ Wrong")

print(f"Found {secret} in {attempts} attempts")
````

````python
# while with continue — skipping iterations

# Print only odd numbers from 1 to 10
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue        # skip even numbers, jump back to the while check
    print(n, end=" ")   # 1 3 5 7 9
````

````python
# while with else — something JavaScript does NOT have!
# The else block runs ONLY if the loop finished naturally (no break)

# Example: searching for a value
numbers = [2, 4, 6, 8, 10]
target = 7
i = 0

while i < len(numbers):
    if numbers[i] == target:
        print(f"Found {target} at index {i}")
        break
    i += 1
else:
    # This runs because the loop ended without hitting break
    print(f"{target} was not found in the list")

# Now try changing target to 6 and run again — the else won't execute!
````

````python
# Processing a list with while — popping items one by one
# Useful when the list changes size during the loop

tasks = ["send email", "write report", "fix bug", "review PR"]

print("Processing tasks:")
while tasks:                      # an empty list is "falsy" (same idea as JS)
    current = tasks.pop(0)        # remove and return the first item
    print(f"  ✓ Done: {current}")

print(f"All done! {len(tasks)} tasks remaining.")
````
